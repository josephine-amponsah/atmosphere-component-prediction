# STATISTICAL ANALYSIS (ANOVA)

From the visualizations, the recorded values of the target variable seem to vary over time. To assess the statistical significance of this variation, all three variables will be tested with ANOVA and the p-value obtain will provide evidence.

```python
#oberved mean variance

def observedMV(var,df):
        df = df[[var]]
        mean = []
        variance = []
        df = df.groupby(df.index.month)
        means = df[var].mean()
        vars = df[var].mean().var()
        mean.append(means)
        variance.append(vars)
        return {'means':mean, 'variance':variance}
```

```python
var = ['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']
```

```python
parameters = {var[i] : observedMV(var[i], data) for i in range(len(var))}

    
```

```python
parameters
```

```python
observed_parameters = pd.DataFrame(parameters)
observed_parameters
```

```python
#function for permuatation test
#var - target variable
#val - index(month  )
def permtest(df,var):
    df[var] = np.random.permutation(df[var])
    variance = df.groupby(df.index.month)[var].mean().var()
    return variance
```

Stat significance of target_carbon_monoxide

```python
carbon_perm_variance = [permtest(data, 'target_carbon_monoxide') for _ in range(1000)]
```

```python
carbon_obs_variance = observed_parameters['target_carbon_monoxide']['variance'][0]
carbon_obs_variance
```

```python
carbon_perm_variance = pd.DataFrame(carbon_perm_variance)
sns.histplot(data = carbon_perm_variance,  bins = 10)
plt.vlines(x= carbon_obs_variance, ymin = 0, ymax =250, colors='blue')
```

```python
pvalue_carbon = np.mean(carbon_perm_variance > carbon_obs_variance)
pvalue_carbon
```

```python
benzene_perm_variance = [permtest(data, 'target_benzene') for _ in range(1000)]
benzene_obs_variance = observed_parameters['target_benzene']['variance'][0]
benzene_perm_variance = pd.DataFrame(benzene_perm_variance)
sns.histplot(data = benzene_perm_variance,  bins = 10)
plt.vlines(x= benzene_obs_variance, ymin = 0, ymax =250, colors='blue')
```

```python
pvalue_benzene = np.mean(benzene_perm_variance > benzene_obs_variance)
pvalue_benzene
```

```python
nitrogen_perm_variance = [permtest(data, 'target_nitrogen_oxides') for _ in range(1000)]
nitrogen_obs_variance = observed_parameters['target_nitrogen_oxides']['variance'][0]
nitrogen_perm_variance = pd.DataFrame(nitrogen_perm_variance)
sns.histplot(data = nitrogen_perm_variance,  bins = 10)
plt.vlines(x= nitrogen_obs_variance, ymin = 0, ymax =250, colors='blue')
```

```python
pvalue_nitrogen = np.mean(nitrogen_perm_variance > nitrogen_obs_variance)
pvalue_nitrogen
```

From the obtained p-values, the null hypothesis cannot be rejected in the case of carbon-monoxide and benzene but can be rejected in the case of nitrogen.

# MULTI TARGET REGRESSION MODEL

```python
#X, y split
X = data.drop(['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides'], axis=1)
y = data[['target_carbon_monoxide', 'target_benzene', 'target_nitrogen_oxides']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Decision Tree model

```python
#model fitting
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor()
model.fit(X_train, y_train)
```

```python
#prediction of y
y_pred = model.predict(X_test)
```

```python
#evaluation of model
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred))
RMSE
```

```python
r2_score(y_test, y_pred)
```

```python
y_pred
```

Random forest regressor

```python
from sklearn.ensemble import RandomForestRegressor
rfmodel = RandomForestRegressor()
```

```python
def run_fmodel(model,X_train,y_train,X_test,y_test):
    
    # Fit Model
    model.fit(X_train,y_train)
    
    # Get Metrics
    
    y_preds = model.predict(X_test)
    
    rmse = np.sqrt(mean_squared_error(y_test,y_preds))
    r2 = r2_score(y_test, y_preds)
    print(f'RMSE : {rmse}')
    print(f'R2 : {r2}')
    
```

```python

for n in range(50,500,50):
    
    model = RandomForestRegressor(n_estimators=n, max_depth= 50)
    
    run_fmodel(model,X_train,y_train,X_test,y_test)
```

With an n_estimater of 200 and maxdepth at 50, the best rmse and r2 scores are obtained. Thus, this will be the value of our hyperparameters for the final model.

```python
model_final = RandomForestRegressor(n_estimators=200, max_depth=50).fit(X_train, y_train)
```

```python
y_pred2 = model_final.predict(X_test)
```

```python
r2_score(y_test, y_pred2)
```

The Random forest regression model gave a significantly higher r2 score than the decision tree model. Thus, the preferred model is the 'rfmodel'. This will thus be applied to the test data for pre diction of the final result.
